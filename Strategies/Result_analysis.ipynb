```python
import pandas as pd
import pymysql
import matplotlib.pyplot as plt 
```

```python
conn = pymysql.connect('localhost', 'root', 'password', 'stock')
cursor = conn.cursor()
sql_1 = """
    SELECT 
  distinct trade_date
FROM
    stock.industry_daily
order by trade_date desc
limit 60;"""
dates = pd.read_sql(sql_1, con=conn)
sql_2 = """
    SELECT 
  *
FROM
    stock.industry_daily
where trade_date >= %s;"""%(dates['trade_date'].values[-1])
df = pd.read_sql(sql_2, con=conn)
```

```python
industry_name = df.drop_duplicates('industry_name')

result = pd.DataFrame()
for i in range(industry_name.shape[0]):
    temp1 = df[df['industry_name'] == industry_name['industry_name'].values[i]]
    std_value = temp1['point'].std()
    mean_value = temp1['point'].mean()
    low3 = 1-3*std_value/mean_value
    low2 = 1-2*std_value/mean_value
    low1 = 1-1*std_value/mean_value
    high3 = 1+3*std_value/mean_value
    high2 = 1+2*std_value/mean_value
    high1 = 1+1*std_value/mean_value
    temp2 = {
        "date":temp1['trade_date'].values[0],
        "name":temp1['industry_name'].values[0],
        "point":temp1['point'].values[0]/mean_value,
        "离散系数":std_value/mean_value,    
        "-3signa":low3,
        "-2signa":low2,
        "-signa":low1,
        "signa":high1,
        "2signa":high2,
        "3signa":high3}
    temp3 = pd.DataFrame(temp2,index = [i])
    result = result.append(temp3)
```

```python
result = result.sort_index(axis = 0, ascending = True, by=['离散系数'])
```
